```python
from pandas_datareader import data
import numpy as np
import pandas as pd
import os
```

```python
oil_price = pd.read_csv('Cushing_OK_WTI_Spot_Price_FOB_1.csv', delimiter = ',', header=4, nrows=1256)
oil_price['Day'] =  pd.to_datetime(oil_price['Day'], format='%m/%d/%Y')
oil_price = oil_price.sort_values('Day')
oil_price.head()
```

```python
qqq = pd.read_csv('daily_QQQ.csv', delimiter=',', skiprows=range(1,6), nrows=1259)
qqq['timestamp'] =  pd.to_datetime(qqq['timestamp'], format='%Y-%m-%d')
qqq = qqq.sort_values('timestamp')
qqq.head()
# print(qqq.info())
```

```python
oil_price_average = oil_price['PriceDollarsPerBarrel'].mean()
oil_price = oil_price.assign(weight=oil_price["PriceDollarsPerBarrel"]/oil_price_average)
oil_price.head()
```

```python
data = qqq.merge(oil_price, left_on='timestamp', right_on='Day', how='inner')
data.head()
```
